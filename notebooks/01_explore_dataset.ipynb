# NeuralVortex — explore a generated HDF5 dataset

Phase-1 sanity notebook. Load the smoke-test HDF5, inspect the schema, and plot a velocity slice + a pressure slice for the first sample.

Run the smoke test first:

```bash
bash scripts/smoke_test.sh   # writes data/smoke.h5
```

````python
import os, sys
import h5py
import numpy as np
import matplotlib.pyplot as plt

if '..' not in sys.path:
    sys.path.insert(0, '..')

PATH = '../data/smoke.h5'
assert os.path.exists(PATH), f'run scripts/smoke_test.sh first; missing {PATH}'

f = h5py.File(PATH, 'r')
print('schema:', dict(f.attrs))
samples = sorted(k for k in f.keys() if k.startswith('sample_'))
print('n_samples =', len(samples))
````

````python
s = f[samples[0]]
print('sample attrs:', dict(s.attrs))
vel = s['velocity'][...]
pres = s['pressure'][...]
print('velocity shape:', vel.shape, 'dtype:', vel.dtype)
print('pressure shape:', pres.shape, 'dtype:', pres.dtype)
print('|v| range:', float(np.linalg.norm(vel, axis=0).min()), '..', float(np.linalg.norm(vel, axis=0).max()))
print('p range:', float(pres.min()), '..', float(pres.max()))
````

````python
G = vel.shape[1]
mid = G // 2
speed = np.linalg.norm(vel, axis=0)
fig, axes = plt.subplots(1, 2, figsize=(10, 4))
im0 = axes[0].imshow(speed[mid, :, :].T, origin='lower', cmap='viridis')
axes[0].set_title(f'|v| (x={mid}-slice)  [m/s]')
plt.colorbar(im0, ax=axes[0], fraction=0.046, pad=0.04)
im1 = axes[1].imshow(pres[mid, :, :].T, origin='lower', cmap='coolwarm')
axes[1].set_title(f'p (x={mid}-slice)  [Pa]')
plt.colorbar(im1, ax=axes[1], fraction=0.046, pad=0.04)
plt.tight_layout()
plt.show()
````
